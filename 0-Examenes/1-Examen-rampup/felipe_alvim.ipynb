2. Escribe una funci칩n en Python que verifique si una palabra o una frase es un pal칤ndromo. Un pal칤ndromo es una palabra o frase que se lee igual de izquierda a derecha que de derecha a izquierda, ignorando los espacios, may칰sculas, min칰sculas y signos de puntuaci칩n.


```python
import re
from unicodedata import normalize
import string

def es_un_palindromo(palabra:str)->str:

    palabra_insertada = palabra.lower().replace(" ","")

    palabra_insertada = re.sub(
        r"([^n\u0300-\u036f]|n(?!\u0303(?![\u0300-\u036f])))[\u0300-\u036f]+", r"\1", normalize("NFD", palabra_insertada),0,re.I
    )
    
    palabra_insertada = palabra_insertada.translate(str.maketrans("","", string.punctuation))

    if palabra_insertada == palabra_insertada[::-1]:
        print('S칤, es un palindromo')
    else:
        print('No es un palindromo')

es_un_palindromo("Dabale arroz, a la Zorr치 el Abad!!!")
```

5. Crea todas las funciones de una calculadora (suma, resta, multiplicaci칩n, divisi칩n)

```python
def calc_suma(x:float, y:float)->float:
    result = x + y
    return result

def calc_subtraccion(x:float, y:float)->float:
    result = x - y
    return result

def calc_multiplicaccion(x:float, y:float)->float:
    result = x * y
    return result

def calc_division(x:float, y:float)->float:
    if y != 0:
        result = x / y
    else:
        print('No se puede dividir por 0')
    return result

```

```python
print(calc_suma(2,3))
print(calc_subtraccion(3,2))
print(calc_multiplicaccion(4,4))
print(calc_division(6,2))
```

6. Crea una clase Calculadora (Opcional)

```python
class Calculadora:

    def __init__(self):
        pass

    def add(self, x:float, y:float)->float:
        result = x + y
        return result


    def subtract(self, x:float, y:float)->float:
        result = x - y
        return result


    def multiply(self, x:float, y:float)->float:
        result = x * y
        return result


    def divide(self, x:float, y:float)->float:
        result = x / y
        return result
```

```python
cal = Calculadora()

cal.add(7,8)
cal.subtract(8,1)
cal.multiply(2,3)
cal.divide(4,2)
```
