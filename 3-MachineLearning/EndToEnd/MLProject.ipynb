```python
import pandas as pd
import numpy as np
import random

def generate_heart_disease_dataset(num_rows=1200):
    """Generates a synthetic heart disease dataset."""

    np.random.seed(42)  # For reproducibility

    ages = np.random.randint(28, 77, num_rows)
    sexes = np.random.choice(['M', 'F'], num_rows)
    chest_pain_types = np.random.choice(['TA', 'ATA', 'NAP', 'ASY'], num_rows)
    resting_bps = np.random.randint(90, 201, num_rows)
    cholesterols = np.random.randint(120, 551, num_rows)
    fasting_bss = np.random.choice([0, 1], num_rows, p=[0.8, 0.2])  # 80% 0, 20% 1
    resting_ecgs = np.random.choice(['Normal', 'ST', 'LVH'], num_rows)
    max_hrs = np.random.randint(70, 201, num_rows)
    exercise_anginas = np.random.choice(['N', 'Y'], num_rows)
    oldpeaks = np.random.uniform(0, 6, num_rows)
    st_slopes = np.random.choice(['Up', 'Flat', 'Down'], num_rows)

    heart_diseases = []
    for i in range(num_rows):
        # Simplified logic to simulate heart disease risk
        risk = 0
        if ages[i] > 55:
            risk += 0.3
        if resting_bps[i] > 140:
            risk += 0.2
        if cholesterols[i] > 200:
            risk += 0.1
        if fasting_bss[i] == 1:
            risk += 0.2
        if exercise_anginas[i] == 'Y':
            risk += 0.3
        if st_slopes[i] == 'Flat' or st_slopes[i] == 'Down':
            risk += 0.4

        if random.random() < risk:
            heart_diseases.append(1)  # Heart disease present
        else:
            heart_diseases.append(0)  # Heart disease absent

    data = {
        'Age': ages,
        'Sex': sexes,
        'ChestPainType': chest_pain_types,
        'RestingBP': resting_bps,
        'Cholesterol': cholesterols,
        'FastingBS': fasting_bss,
        'RestingECG': resting_ecgs,
        'MaxHR': max_hrs,
        'ExerciseAngina': exercise_anginas,
        'Oldpeak': oldpeaks,
        'ST_Slope': st_slopes,
        'HeartDisease': heart_diseases
    }

    abc = pd.DataFrame(data)
    return abc

# Generate and save the dataset to a CSV file
heart_disease_abc = generate_heart_disease_dataset()
heart_disease_abc.to_csv('heart_disease_dataset.csv', index=False)

print("Dataset generated and saved to 'heart_disease_dataset.csv'")
print(heart_disease_abc.head())
print(heart_disease_abc.shape)
```
