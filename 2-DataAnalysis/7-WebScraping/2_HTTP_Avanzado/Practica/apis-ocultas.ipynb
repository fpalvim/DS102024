```python
base_url = "https://www.idealista.com"
target_url = "/venta-viviendas/rivas-vaciamadrid/rivas-futura/"

descripcion_piso = []

while target_url:

    resp = requests.get(base_url + target_url, headers=headers)
    soup = BeautifulSoup(resp.text, 'html.parser')

    todos_pisos = soup.find_all("div", {"class": "item-info-container"})

    for piso in todos_pisos:
        etiquetas_infos_piso = {
            "Título": piso.find("a", {"class": "item-link"}).text.strip("\n"),
            "Precio": piso.find("span", {"class": "item-price"}).text.strip("\n"),
            "Área": piso.find("div", {"class": "item-detail-char"}).text.strip("\n"),
            "Descripción": piso.find("div", {"class": "item-description"}).text.strip("\n"),
            "Enlace": base_url + piso.find("a", {"class": "item-link"}).get('href')
        }
        descripcion_piso.append(etiquetas_infos_piso)

    next_page = soup.find("a", {"class": "icon-arrow-right-after"})
    
    target_url = next_page['href'] if next_page else None

print(descripcion_piso)
```

```python
df_pisos = pd.DataFrame(descripcion_piso, columns=["Título", "Precio", "Área", "Descripción", "Enlace"])
```
