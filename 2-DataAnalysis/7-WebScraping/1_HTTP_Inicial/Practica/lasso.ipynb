Descargar letras de canciones

Utilizando beautiful soup descargar todas las canciones de [Lasso](https://es.wikipedia.org/wiki/Lasso_(cantante)) que hay en [letras.com](https://www.letras.com/lasso/)

```python
from bs4 import BeautifulSoup
import requests
import os
import pandas as pd



def descargar_letras(artista):

    letras_url = f"https://www.letras.com/{artista}"

    url_obtenido = requests.get(letras_url)
    soup = BeautifulSoup(url_obtenido.text, "html.parser")
    
    songs_container = soup.find_all('li', class_='songList-table-row --song isVisible')

    songs_list = []

    for song in songs_container:
        songs_list.append((song['data-name'][0], song['data-name']))

    df = pd.DataFrame(songs_list, columns=['Inicial','Titulo'])

    return df
```

```python
descargar_letras('charlie-brown-jr')
```
