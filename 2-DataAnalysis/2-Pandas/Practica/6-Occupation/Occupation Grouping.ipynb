### Paso 4. Descubre cuál es la edad media por ocupación

```python
round(usuarios.groupby('occupation')[['age']].mean(),2)
```

### Paso 5. Descubra la proporción de Hombres por ocupación y ordénela de mayor a menor

#### Forma 1

```python
usuarios.groupby(['occupation'])['gender'].value_counts(normalize=True).xs('M', level='gender')*100
```

#### Forma 2

```python
usuarios.groupby(['occupation','gender'])['gender'].value_counts().unstack()
```

```python
df_people_by_role = usuarios.groupby('occupation')['gender'].value_counts().unstack(fill_value=0.0)
total_both_by_role = df_people_by_role['M'] + df_people_by_role['F']
total_men_by_role = (df_people_by_role['M'] / total_both_by_role) * 100
total_men_by_role
```

### Paso 6. Para cada ocupación, calcule las edades mínima y máxima

```python
usuarios.groupby('occupation')[['age']].agg(['min','max'])
```

### Paso 7. Para cada combinación de ocupación y sexo, calcule la edad media

```python
usuarios.groupby(['occupation','gender'], as_index=False)['age'].mean()
```

### Paso 8.  Para cada ocupación, presente el porcentaje de mujeres y hombres

```python
merged_usuarios = usuarios.groupby('occupation')['gender'].value_counts().unstack(fill_value=0.0)
merged_usuarios['total'] = merged_usuarios['F'] + merged_usuarios['M']
merged_usuarios['women %'] = (merged_usuarios['F'] / merged_usuarios['total'])*100
merged_usuarios['men %'] = (merged_usuarios['M'] / merged_usuarios['total'])*100
merged_usuarios
```

```python
a = usuarios.query('gender == "F"').value_counts().sum()
b = usuarios.query('gender == "M"').value_counts().sum()

tot = a + b
perc_a = (a/tot)*100
perc_a
```
